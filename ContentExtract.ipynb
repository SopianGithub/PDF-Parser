```python
import re
from PDFContentConverter import PDFContentConverter

converter = PDFContentConverter('samples/Contents.pdf')

#
comNumRegex = re.compile(r'[A-z]-[0-9]')

# result = converter.pdf2pandas()

df = converter.pdf2pandas()
# Remove Whitespace
df['text'] = df['text'].str.replace('.', '')
df['text'] = df['text'].str.strip()
# Get Content Text with Lenth of Char
df['text'] = df['text'].astype('str')
mask = (df['text'].str.len() > 10)
df = df.loc[mask]
# Exlude Content Header
excText = df[ (df['text'].str.strip() == 'MEDCOM Pam 40-7-21') ].index
df.drop(excText , inplace=False)

# print(df)

new_df = df[['id','text']].copy()
new_df['text_check'] = new_df['text'].str.match(comNumRegex)
new_df['new_text'] = new_df['text'].replace(regex=r'[A-z]-[0-9]', value='')

print(new_df)
```
